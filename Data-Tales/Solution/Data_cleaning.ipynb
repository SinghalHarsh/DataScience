```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
train = pd.read_csv('train.csv', parse_dates=['Day.Index'], dayfirst=True)
test = pd.read_csv('test.csv', parse_dates=['Day.Index'], dayfirst=True)
ss = pd.read_csv('samplesubmission.csv')
```

```python
#Extracting Day, month, weekday, year from Date
```

```python
train['Day.Index'] = pd.to_datetime(train['Day.Index'], format='%y-%m-%d')
test['Day.Index'] = pd.to_datetime(test['Day.Index'], format='%y-%m-%d')

train['Year'] = train['Day.Index'].dt.year
train['Month'] = train['Day.Index'].dt.month
train['Week'] = train['Day.Index'].dt.week
train['Day_Of_Week'] = train['Day.Index'].dt.dayofweek
train['Day'] = train['Day.Index'].dt.day
train['Day_Of_Year'] = train['Day.Index'].dt.dayofyear

test['Year'] = test['Day.Index'].dt.year
test['Month'] = test['Day.Index'].dt.month
test['Week'] = test['Day.Index'].dt.week
test['Day_Of_Week'] = test['Day.Index'].dt.dayofweek
test['Day'] = test['Day.Index'].dt.day
test['Day_Of_Year'] = test['Day.Index'].dt.dayofyear
```

```python
#Checking Null values
```

```python
train.isnull().any()

test.isnull().any()
```

```python
data = train.append(test, ignore_index=True)
data['Obtained.Leads'].fillna(0, inplace=True)
```

```python
def function1(X):
    return int(round(X["Amount.Spend"]/X["Obt.Click"]))
```

```python
data["CPC"]=data.apply(function1,axis=1)
train["CPC"]=train.apply(function1,axis=1)
test["CPC"]=test.apply(function1,axis=1)
```

```python
train.to_csv("modified_train.csv",index=False)
test.to_csv("modified_test.csv",index=False)
data.to_csv('modified_data.csv',index=False)
```
