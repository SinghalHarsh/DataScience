```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
from matplotlib import rcParams
#colorbrewer2 Dark2 qualitative color table
dark2_colors = [(0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843)]

rcParams['figure.figsize'] = (8, 3)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'
rcParams['axes.grid'] = True
rcParams['axes.facecolor'] = '#eeeeee'
```

```python
train = pd.read_csv('modified_train.csv', parse_dates=['Day.Index'], dayfirst=True)
test = pd.read_csv('modified_test.csv', parse_dates=['Day.Index'], dayfirst=True)
data = pd.read_csv('modified_data.csv', parse_dates=['Day.Index'], dayfirst=True)
```

```python
data.head()
```

```python
#Univariate_Analysis
```

```python
def cont_plot(X):
    rcParams['figure.figsize'] = (8, 6)
    fig, arr = plt.subplots(2,2)
    sns.distplot(train[X],ax=arr[0,0])
    arr[0,0].set_title('Train')
    sns.distplot(test[X],ax=arr[0,1])
    arr[0,1].set_title('Test')
    sns.distplot(data[X],ax=arr[1,0])
    arr[1,0].set_title('Data')
```

```python
def cate_plot(X):
    rcParams['figure.figsize'] = (8, 3)
    fig, (ax1, ax2, ax3) = plt.subplots(1,3)
    sns.countplot(X,data=train,ax=ax1,)
    ax1.set_title('Train')
    sns.countplot(X,data=test,ax=ax2)
    ax2.set_title('Test')
    sns.countplot(X,data=data,ax=ax3)
    ax3.set_title('Data')
```
